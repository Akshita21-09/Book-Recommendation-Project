```python
# Copying the manipulated data for df2 for backup
# Resting index to allign with the model
# Creating rate categories 0-1, 1-2, 2-3, 3-4, 4-5
# One-Hot encoding - converting categorical variables into numeric format
# Creating feature matrix including rating category,language,average_rating,ratings_count
```

```python
df2=df.copy()
df2=df2.reset_index(drop=True)
df2.loc[(df2['average_rating']>=0)&(df2['average_rating']<=1),'rating between']='between 0 and 1'
df2.loc[(df2['average_rating']>1)&(df2['average_rating']<=2),'rating between']='between 1 and 2'
df2.loc[(df2['average_rating']>2)&(df2['average_rating']<=3),'rating between']='between 2 and 3'
df2.loc[(df2['average_rating']>3)&(df2['average_rating']<=4),'rating between']='between 3 and 4'
df2.loc[(df2['average_rating']>4)&(df2['average_rating']<=5),'rating between']='between 4 and 5'
df2.sort_values('average_rating')
rating_df=pd.get_dummies(df2['rating between'])
language_df=pd.get_dummies(df2['language_code'])
feature=pd.concat([rating_df,language_df,df2['average_rating'],df2['ratings_count']],axis=1)
```

```python
# importing MinMaxScaler,neighbors,train_test_split from sklearn
# Normalizing the features
# Creating KNN Model - n_neighbors=8 -> for returning 8 similar books
# ball_tree -> efficient nearest neighbor algorithm
# Fitting the model - Trains model using feature matrix
# finding neighbors - dist -> distances, idlist -> indices of nearest neighbors
# defining functions for book recommendation - finds book index, get its nearest neighbor indices, returns corresponding titles
# returns 8 similar books for book given
```

```python
from sklearn.preprocessing import MinMaxScaler
minmax=MinMaxScaler()
feature=minmax.fit_transform(feature)
from sklearn import neighbors
from sklearn.model_selection import train_test_split
model=neighbors.NearestNeighbors(n_neighbors=8,algorithm='ball_tree')
model.fit(feature)
dist,idlist=model.kneighbors(feature)
def bookRecom(name):
    book_list=[]
    book_id=df2[df2['title']==name].index
    book_id=book_id[0]
    for new in idlist[book_id]:
        book_list.append(df2.loc[new].title)
    return book_list
Book_name=bookRecom('Harry Potter and the Half-Blood Prince (Harry Potter  #6)')
Book_name

```

```python
from sklearn.neighbors import NearestNeighbors

model = NearestNeighbors(metric='cosine')
model.fit(feature)
dist, idlist = model.kneighbors(feature)

def bookRecom(name, n=5):
    matches = df2[df2['title'].str.contains(name, case=False)]
    
    if matches.empty:
        return "Book not found in dataset"
    
    book_id = matches.index[0]
    
    recommendations = []
    
    for i, new in enumerate(idlist[book_id][1:n+1]):
        recommendations.append({
            "Title": df2.loc[new].title,
            "Author": df2.loc[new].authors,
            "Average Rating": df2.loc[new].average_rating,
            "Similarity Score": round(1 - dist[book_id][i+1], 3)
        })
    
    return pd.DataFrame(recommendations)
```

```python
def bookRecom(name, n=5):
    matches = df2[df2['title'].str.contains(name, case=False)]
    
    if matches.empty:
        return "Book not found in dataset"
    
    book_id = matches.index[0]
    
    recommendations = []
    
    for i, new in enumerate(idlist[book_id][1:n+1]):
        recommendations.append(
            f"{df2.loc[new].title} | ⭐ {df2.loc[new].average_rating}"
        )
    
    return "\n".join(recommendations)
```
